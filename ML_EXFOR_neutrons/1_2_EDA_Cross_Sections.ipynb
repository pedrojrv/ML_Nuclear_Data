```python
groups = uranium_233_mt2[["Energy", "Data", "Short_Reference"]].groupby("Short_Reference")
plt.figure(figsize=(14,10))
ax = plt.subplot(111)
ax.plot(uranium_endf.Energy, uranium_endf.Data, label="ENDF/B-VIII.0", alpha=0.5)
for name, group in groups:
    ax.plot(group["Energy"], group["Data"], marker="o", linestyle="", label=name, alpha=0.9)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("Energy (eV)")
ax.set_ylabel("Cross Section (b)")

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend(fontsize=21, loc=3)
plt.savefig(os.path.join(figure_dir, "EXFOR_{}_{}_XS.png".format(uranium_233_mt2.Isotope.values[0], "2")), bbox_inches='tight', dpi=600)
```
