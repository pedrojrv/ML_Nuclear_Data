```python
import pandas as pd
import numpy as np
```

# Parsing Atomic Masses, Mass Excess, Binding Energy, and Beta Decay Energy.

```python
with open("mass16_toparse.txt") as infile, open('atomic_masses.txt', 'w') as outfile:
    for line in infile:
        if line.strip():
            string = list(line)
            for i, j in enumerate([1, 4, 9, 14, 19, 22, 27, 41, 52, 63, 72, 75, 86, 95, 112]):
                string.insert(i + j, ',')
            outfile.write("".join(string))
```

```python
columns = ["Page_Feed", "NZ", "N", "Z", "A", "EL", "O", \
           "Mass_Excess", "dMass_Excess", "Binding_Energy", "dBinding_Energy", \
           "Beta_Type", "B_Decay_Energy", "dB_Decay_Energy", "Atomic_Mass_Micro", "dAtomic_Mass_Micro"]
```

```python
df = pd.read_csv("atomic_masses.txt", header=None, names=columns)
df = df.drop(columns=["Page_Feed", "NZ"])
```

```python
df.head()
```

```python
strings = ["EL", "O", "Mass_Excess", "dMass_Excess", "Binding_Energy", "dBinding_Energy", \
           "Beta_Type", "B_Decay_Energy", "dB_Decay_Energy", "Atomic_Mass_Micro", "dAtomic_Mass_Micro"]
```

```python
for col in strings:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.strip("\"")
    df[col] = df[col].str.strip()
```

```python
# for col in strings:
#     print("Value Counts for {} column.".format(col))
#     print(df[col].value_counts())
#     print("\n")
```

```python
df['dB_Decay_Energy'].loc[df['B_Decay_Energy'] == '?'] = "?"
```

```python
df.head()
```

```python
df["O"] = df["O"].replace(to_replace="", value="Other")
```

```python
df = df.drop(columns="Beta_Type")
```

```python
df = df.replace('?', np.nan)
```

```python
df.head()
```

```python
df["Atomic_Mass_Micro"] = df['Atomic_Mass_Micro'].str.replace(" ", "")
```

```python
for col in df.columns[5:]:
    df[col] = df[col].astype(float)
```

```python
df.info()
```

```python
df.to_csv("atomic_masses.csv", index=False)
```

# Parsing Separation Energy Data

```python
with open("rct1_16.txt") as infile, open('rct1_16_sep.txt', 'w') as outfile:
    for line in infile:
        if len(line) != 120:
                line = line[:-1]
                spaces = 120 - len(line)
                line = line + (" " * spaces)  + "\n"
                
        if line.strip():
            string = list(line)
            for i, j in enumerate([1, 4, 8, 11, 22, 30, 40, 48, 58, 66, 76, 84, 94, 102, 112]):
                string.insert(i + j, ',')
            outfile.write("".join(string))
```

```python
with open("rct2_16.txt") as infile, open('rct2_16_sep.txt', 'w') as outfile:
    for line in infile:
        if len(line) != 120:
                line = line[:-1]
                spaces = 120 - len(line)
                line = line + (" " * spaces)  + "\n"
                
        if line.strip():
            string = list(line)
            for i, j in enumerate([1, 4, 8, 11, 22, 30, 40, 48, 58, 66, 76, 84, 94, 102, 112]):
                string.insert(i + j, ',')
            outfile.write("".join(string))
```

```python
columns_rct1 = ["Page_Feed", "A", "EL", "Z", \
           "S(2n)", "dS(2n)", "S(2p)", "dS(2p)", "Q(a)",  "dQ(a)", "Q(2B-)", "dQ(2B-)", "Q(ep)", "dQ(ep)", "Q(B-n)", "dQ(B-n)"]
rct1 = pd.read_csv("rct1_16_sep.txt", header=None, names=columns_rct1)
```

```python
rct1.head()
```

```python
columns_rct2 = ["Page_Feed", "A", "EL", "Z", \
           "S(n)", "dS(n)", "S(p)", "dS(p)", "Q(4B-)", "dQ(4B-)", "Q(d,a)", "dQ(d,a)", "Q(p,a)", "dQ(p,a)", "Q(n,a)", "dQ(n,a)"]
rct2 = pd.read_csv("rct2_16_sep.txt", header=None, names=columns_rct2)
```

```python
rct2.head()
```

```python
for col in columns_rct1:
    rct1[col] = rct1[col].astype(str)
    rct1[col] = rct1[col].str.strip("\"")
    rct1[col] = rct1[col].str.strip()
    rct1[col] = rct1[col].str.replace("#", ".")
    
for col in columns_rct2:
    rct2[col] = rct2[col].astype(str)
    rct2[col] = rct2[col].str.strip("\"")
    rct2[col] = rct2[col].str.strip()
    rct2[col] = rct2[col].str.replace("#", ".")
```

```python
rct1 = rct1.replace(to_replace="*", value=np.nan)
rct2 = rct2.replace(to_replace="*", value=np.nan)

rct1 = rct1.replace(to_replace="", value=np.nan)
rct2 = rct2.replace(to_replace="", value=np.nan)
```

```python
for col in list(rct1.columns)[4:]:
    rct1[col] = rct1[col].astype(float)
    
for col in list(rct2.columns)[4:]:
    rct2[col] = rct2[col].astype(float)
```

```python
rct1[["A", "Z"]] = rct1[["A", "Z"]].astype(int)
rct2[["A", "Z"]] = rct2[["A", "Z"]].astype(int)

rct1.drop(columns=["Page_Feed"], inplace=True)
rct2.drop(columns=["Page_Feed"], inplace=True)
```

```python
rct1.head()
```

```python
rct2.head()
```

```python
rct1["N"] = rct1["A"] - rct1["Z"]
rct2["N"] = rct2["A"] - rct2["Z"]
```

# Merging All Three AMES datafiles

```python
pd.set_option('display.max_columns', 500)
```

```python
new_df = pd.merge(df, rct1,  how='left', left_on=['N','Z'], right_on = ['N','Z'])
new_df = pd.merge(new_df, rct2,  how='left', left_on=['N','Z'], right_on = ['N','Z'])
```

```python
new_df
```

# Adding More Reaction Energies (2017Wang10)

```python
new_df["Q(g,p)"] = -1 * new_df["S(p)"]
new_df["Q(g,n)"] = -1 * new_df["S(n)"]
new_df["Q(g,pn)"] = new_df["Q(d,a)"] - 26071.0939
new_df["Q(g,d)"] = new_df["Q(d,a)"] - 23846.5279
new_df["Q(g,t)"] = new_df["Q(p,a)"] - 19813.8649
new_df["Q(g,He3)"] = new_df["Q(n,a)"] - 20577.6194
new_df["Q(g,2p)"] = -1 * new_df["S(2p)"]
new_df["Q(g,2n)"] = -1 * new_df["S(2n)"]
new_df["Q(g,a)"] = new_df["Q(a)"]
new_df["Q(p,n)"] = new_df["B_Decay_Energy"] - 782.3465
new_df["Q(p,2p)"] = -1 * new_df["S(p)"]
new_df["Q(p,pn)"] = -1 * new_df["S(n)"]
new_df["Q(p,d)"] = -1 * new_df["S(n)"] + 2224.5660
new_df["Q(p,2n)"] = new_df["Q(B-n)"] - 782.3465
new_df["Q(p,t)"] = -1 * new_df["S(2n)"] + 8481.7949
new_df["Q(p,3He)"] = new_df["Q(d,a)"] - 18353.0535
new_df["Q(n,2p)"] = new_df["Q(ep)"] + 782.3465
new_df["Q(n,np)"] = -1 * new_df["S(p)"]
new_df["Q(n,d)"] = -1 * new_df["S(p)"] + 2224.5660
new_df["Q(n,2n)"] = -1 * new_df["S(n)"]
new_df["Q(n,t)"] = new_df["Q(d,a)"] - 17589.2989
new_df["Q(n,3He)"] = -1 * new_df["S(2p)"] + 7718.0404
new_df["Q(d,t)"] = -1 * new_df["S(n)"] + 6257.2290
new_df["Q(d,3He)"] = -1 * new_df["S(p)"] + 5493.4744
new_df["Q(3He,t)"] = new_df["B_Decay_Energy"] - 18.5920
new_df["Q(3He,a)"] = -1 * new_df["S(n)"] + 20577.6194
new_df["Q(t,a)"] = -1 * new_df["S(p)"] + 19813.8649
```

```python
drop_redundant = ["A_x", "EL_x", "A_y", "EL_y"]
new_df = new_df.drop(columns=drop_redundant)
```
