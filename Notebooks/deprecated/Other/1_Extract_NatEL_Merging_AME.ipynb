# Periodic Table

```python
txt_dir = "./AME_Files/"
```

```python
masses_natural = pd.read_csv(txt_dir + "periodic_table.csv").rename(
    columns={'NumberofNeutrons':'Neutrons', 'NumberofProtons':'Protons', 'AtomicMass':'Atomic_Mass_Micro', 'Symbol':'EL'})
```

```python
masses_natural["Mass_Number"] = masses_natural["Neutrons"] + masses_natural["Protons"]
```

```python
masses_natural.head()
```

```python
masses_natural = masses_natural[["Neutrons", "Protons", "Mass_Number", "EL", "Atomic_Mass_Micro", "NumberofElectrons", "AtomicRadius", "Electronegativity", "Density"]]
```

```python
masses_natural.head()
```

```python
masses_natural[masses_natural["Protons"] == 92]
```

```python
masses_natural["N"] = masses_natural["Neutrons"] * 0
```

```python
masses_natural["A"] = masses_natural["Mass_Number"] * 0
```

```python
masses_natural.to_csv("natural_element_masses.csv", index=False)
```

# Merging with AME

```python
natural = pd.read_csv("natural_element_masses.csv")
ame = pd.read_csv("./AME/elemental_properties.csv")
```

```python
natural.columns = ["N", "Z", "A", "EL", "Atomic_Mass_Micro", "Neutrons", "Mass_Number"]
natural["Neutrons"] = natural["Mass_Number"] - natural["Z"]
natural.head()
```

```python
ame.head()
```

```python
result = ame.append(natural, sort=False)
```

```python
result[result["Z"] == 92]
```

```python
result.Neutrons.fillna(result.N, inplace=True)
result.Mass_Number.fillna(result.A, inplace=True)
```

```python
for i in range(0,119):
    result[result["Z"] == i] = result[result["Z"] == i].fillna(result[result["Z"] == i].mean())
```

```python
result.Neutrons = result.Neutrons.astype(int)
result.Mass_Number = result.Mass_Number.astype(int)
```

```python
result["O"].fillna(value="Other", inplace=True)
```

```python
result.fillna(value=0, inplace=True)
```

```python
result.head()
```

```python
result.to_csv("Final_Element_Properties.csv", index=False)
```
