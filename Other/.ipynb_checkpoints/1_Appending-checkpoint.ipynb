```python
import pandas as pd
```

# Using Extracted Data and Appending It

```python
df = pd.read_csv("data_points_per_experiment_refined.txt",  delim_whitespace=True, header=None)
df.columns = ["Data", "Multiple"]
```

```python
df.shape
```

```python
df.head()
```

```python
df1 = pd.read_csv("reaction_notation.txt", delim_whitespace=True, header=None)
df1.columns = ["Reaction", "Type"]
```

```python
df1.shape
```

```python
df1.head()
```

```python
df2 = pd.read_csv("titles.txt", sep="      ", header=None)
df2.columns = ["Keyword", "Title"]
```

```python
df2.shape
```

```python
df2.head()
```

```python
pre_final = pd.concat([df, df1, df2], axis=1)
pre_final.head()
```

```python
final = pre_final.reindex(pre_final.index.repeat(pre_final.Multiple))
final['position'] = final.groupby(level=0).cumcount() + 1
```

```python
final.head()
```

```python
final["reaction_notation"] = final.Type.str.extract('.*\((.*)\).*')
```

```python
final.head()
```

```python
final["reaction_notation"].value_counts()
```

```python
final["reaction_notation2"] = final["reaction_notation"].apply(lambda x: x.split(')')[0])
```

```python
final["reaction_notation2"].value_counts()
```

```python
final = pd.concat([final, final["reaction_notation2"].str.split(',', expand=True)], axis=1)
```

```python
final.head()
```

```python
new_columns = list(final.columns)
new_columns = new_columns[:9]
new_columns
```

```python
new_columns.extend(["Projectile", "Out"])
```

```python
final.columns = new_columns
final.head()
```

```python
final = final[["Type", "Title", "Out"]]
```

```python
final.shape
```

```python
final.head()
```

```python
final.to_csv("reaction_and_titles.csv")
```
