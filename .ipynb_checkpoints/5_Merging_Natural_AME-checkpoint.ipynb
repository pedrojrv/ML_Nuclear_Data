```python
import pandas as pd
```

```python
natural = pd.read_csv("natural_element_masses.csv", index_col=0)
ame = pd.read_csv("atomic_masses.csv", index_col=0)
```

```python
natural.columns = ["N", "Z", "A", "EL", "Atomic_Mass_Micro"]
natural.head()
```

```python
ame.head()
```

```python
result = ame.append(natural, sort=False)
```

```python
result.tail()
```

```python
result[result["Z"] == 92]
```

```python
for i in range(1,119):
    result[result["Z"] == i] = result[result["Z"] == i].fillna(result[result["Z"] == i].mean())
```
